```python
from sklearn.neighbors import KNeighborsClassifier 
```

```python
knn = KNeighborsClassifier(n_neighbors= 1)
```

```python
knn.fit(X_egitim,y_egitim)
```

```python
import numpy as np
X_yeni = np.array([[5,2.9,1,0.2]])
X_yeni.shape
```

```python
tahmin = knn.predict(X_yeni)
print("Tahmin sinif: ",tahmin)
print("Tahmin Turu: ",iris["target_names"][tahmin])
```

```python
y_tahmin = knn.predict(X_test)
print(y_tahmin)
```

```python
print(np.mean(y_tahmin==y_test))
```

```python
print(knn.score(X_test,y_test))
```
